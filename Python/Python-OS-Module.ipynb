The os module provides portable ways to interact with the operating system: paths, directories, environment variables, and low-level file operations.

Working with OS directory and paths

```python
import os
```

# Finding current working directory

```python
print(os.getcwd())
```

# changing current directory

```python
os.chdir('..')
print(os.getcwd())
```

# join paths portably

```python
p = os.path.join('folder', 'subfolder', 'file.txt')
print(p)
```

# Finding an absolute path

```python
print(os.path.abspath(p))
```

# Splitting Path

```python
print('basename:', os.path.basename(p))
print('dirname: ', os.path.dirname(p))
```

# Existing paths

```python
print(os.path.exists(p))
```

# Listing directories

```python
print(os.listdir('.'))
```

# Creating, renaming, deleting files & directories

Creating one directory

```python
os.mkdir('test_dir')
```

Creating nested directories

```python
os.makedirs('test_dir/a/b', exist_ok=True)
```

Renaming a directory

```python
os.rename('test_dir', 'new_test_dir')
```
